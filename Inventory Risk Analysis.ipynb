```python
cols = [ 'Days for shipping (real)','Days for shipment (scheduled)','Sales per customer','Delivery Status','Late_delivery_risk', 'Category Id',
        'Category Name', 'Customer City', 'Customer Country', 'Customer Id','order date (DateOrders)', 'Order Id','Sales','Order Region', 'Order State', 
        'Order Status','Product Name', 'Product Price','Product Status', 'shipping date (DateOrders)', 'Shipping Mode','Order Item Quantity',
        'Order Item Total', 'Order Profit Per Order' ]
#v we dont need all entire datasset we need those with which we needed 
```

```python
data = data[cols].copy()
data.isnull().sum().sum()   # No null values data is clear
```

```python
data.dtypes     # here we can see the date is in integer value lets convert it to date first
```

```python
data["order date (DateOrders)"] = pd.to_datetime(data["order date (DateOrders)"])
```

```python
data["shipping date (DateOrders)"] = pd.to_datetime(data["shipping date (DateOrders)"])
```

Basic Data Manupulation

```python
data.columns = (data.columns
                .str.strip()
                .str.lower()
                .str.replace(" ","_")
                .str.replace(r"[()]","", regex =True ))
data.columns
```

# Insights

Lead Time Variance

```python
data["lead_time_variance"] = data["days_for_shipping_real"] - data["days_for_shipment_scheduled"]

conditions = [ 
    data["lead_time_variance"] == 0,
    data["lead_time_variance"] < 0             
]
choices = ["On-Time","Early"]

data["shipment_status"] = np.select( conditions, choices, default = "Late")

# to get shipment delivery status
```

```python
data.drop(columns = ["product_status"], inplace = True)
data.drop(columns = ["delivery_status"], inplace = True)
```

```python
data.rename(columns ={ "order_profit_per_order": "profit_per_order"}, inplace = True)
data.columns
```

```python
data.rename(columns = {"profit_per_order" : "profit"}, inplace = True)
```

```python
data["profit_margin"] = data["profit"] / data["sales"]
```

Profit Margin %

```python
data["profit_margin_pct"] = (data["profit"] / data["sales"]) * 100
print(f"Profit Margin is :{data['profit_margin_pct']}")
```

## Risk Category (condition-based) — Without ML

```python
condition = [ (data["late_delivery_risk"] == 1) &(data["lead_time_variance"] > 3),
             (data["late_delivery_risk"] == 1) & (data["lead_time_variance"] > 1)]

choice = ["High","Medium"]

data["risk_category"] = np.select(condition,
                                  choice,
                                  default = "Low")
data["risk_category"]
```
